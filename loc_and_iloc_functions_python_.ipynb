# ***iloc and loc functions in python***

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'name':['emmanuel','simon','hero','ironman'],
 'class':[1,2,3,4],
 'sub':['science','maths','kannada','hindi'],
 'dist':['banglore','bidar','gulbarga','huballi']})
```

```python
df
```

```python
syntax:
df.iloc(range of rows index,range of column index)
```

```python
df.iloc[:,:]
# selects all rows and all columns
```

```python
df.iloc[0:2,0:3]
#selects 0 to 2 rows and 0 to 3 columns 
```

```python
df.iloc[1:,2:]
# selects 1 to all rows and 2 to all columns
```

```python
df
```

**df.loc** *
here in df.loc we need to give columns names instead of index numbers 

```python
df
```

```python
df.loc['1':'2','name':'dist']
# here l to 2 rows are selected and name to dist columns are selected
```

```python
df.loc['1':,'name':]
# here 1 to all rows are selected and name to all columns selected 
```
